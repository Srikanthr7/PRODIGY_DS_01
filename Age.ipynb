```python
import matplotlib.pyplot as plt

# Create some data
ages = [20, 30, 40, 50, 60, 70, 80, 90, 100]
genders = ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M']

# Create a bar chart of the ages
plt.bar(ages, [1 for _ in ages])
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Distribution of Ages')
plt.show()

# Create a histogram of the ages
plt.hist(ages)
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Distribution of Ages')
plt.show()

# Create a bar chart of the genders
plt.bar(genders, [1 for _ in genders])
plt.xlabel('Gender')
plt.ylabel('Number of people')
plt.title('Distribution of Genders')
plt.show()
```
